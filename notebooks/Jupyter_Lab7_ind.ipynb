## Индивидуальное задание


Найти изображение в сети Интернет. Сформировать матрицу выбранного изображения, записать ее в файл с расширением png. Изображение, записанное в этом файле, вывести на экран. Получить негатив изображения, вывести на экран. Выделить и вывести на экран определенный регион изображения.


```python
import cv2
import requests
import numpy as np
from matplotlib import pyplot as plt
from io import BytesIO
from PIL import Image


# get request
url = requests.get('https://i.pinimg.com/originals/23/67/c9/2367c9a9a6645995573a0f8209a848e6.png')
# response.content returns the content of the response, in bytes
img_orig = Image.open(BytesIO(url.content))
# Преобразование цветового пространства
img_orig = cv2.cvtColor(np.array(img_orig), cv2.COLOR_RGB2RGBA)

# Запись в файл
cv2.imwrite('img_ind.png', img_orig)
# Вывод
plt.imshow(img_orig);
cv2.imshow('img_ind', img_orig)
cv2.waitKey(0)
cv2.destroyAllWindows ()

# Формирование матрицы цветного изображения
img = cv2.imread("img_ind.png", 1)
# Рисуем прямоугольник, (421,500), (562,690) − координаты левого верхнего и правого нижнего угла; (0,0,255) − красный цвет, 2 − толщина линии.
image=cv2.rectangle(img,(421,500),(562,690),(0,0,255), 2)
cv2.imshow('Gray cat', img)
cv2.waitKey(0)

# Чтение полутонного изображения
img = cv2.imread("img_ind.png", 0)
# монохромное изображение
img = cv2.cvtColor(img, cv2.IMREAD_GRAYSCALE)
# Превратим в серый
img = cv2.bitwise_not(img) # Функция инвертирования изображения
cv2.imshow('Negative',img)
cv2.waitKey(0)
cv2.destroyAllWindows ()
```
